```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
reviews=pd.read_csv("/home/saumitra/Desktop/kaggle/winemag-data-130k-v2.csv",index_col=0)
reviews.head(3)
```

```python
from plotnine import *
```

```python
top_wines=reviews[reviews['variety'].isin(reviews.variety.value_counts().head(5).index)]
top_wines.head(3)
```

# Simple Scatter plot

```python
df=top_wines.head(1000).dropna()
(ggplot(df) + aes(x='points',y='price') + geom_point())
```

**Adding regression line with stat smooth**

```python
df=top_wines.head(1000).dropna()
(ggplot(df) + aes(x='points',y='price') + geom_point() + stat_smooth())
```

**Adding colors**

```python
df=top_wines.head(1000).dropna()
(ggplot(df) + aes(color='points') + aes(x='points',y='price') + geom_point() + stat_smooth()
)
```

**Faceting**

```python
df=top_wines.head(1000).dropna()
(ggplot(df) + aes(color='points') + aes(x='points',y='price') + 
 geom_point() + stat_smooth() + facet_wrap('~variety'))
```

```python
#aes can appear as a layer element
df=top_wines.head(1000).dropna()
(ggplot(df) + geom_point(aes(x='points',y='price')))
```

```python
#aes can appear as a parameter in the overall data
(ggplot(df,aes(x='points',y='price')) + geom_point())
```

# Barplot

```python
(ggplot(top_wines) + aes(x='points') + geom_bar())
```

```python
#Equivalent to hexplot
(ggplot(top_wines) + aes('points' , 'variety') + geom_bin2d(bins=20))
```

```python
#Non-geometric function calls can be mixed in to change the structure of the plot. We've already seen `facet_wrap`;
#`coord_fixed` and `ggtitle` are two more.
(ggplot(top_wines) + aes('points','variety') + geom_bin2d(bins=20) + coord_fixed(ratio=2) +
 ggtitle('Top Five Most Common Wine Variety Points Awarded'))
```
