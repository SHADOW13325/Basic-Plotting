**Attack plotting in histogram**

```python
bins=range(0,200,20)
plt.hist(my_data['ATTACK'],bins,color='yellow')
plt.xlabel('Attack')
plt.ylabel('Count')
plt.plot()
plt.axvline(my_data['ATTACK'].median(),linestyle='dashed',color='red') #making dashed line
plt.show()
```

```python
my_data['TYPE_1'].value_counts()
```

**Comparing two different types of pokemons ---Scatterplot**

```python
plt.subplots(figsize=(12,8))
Psychic=my_data[(my_data['TYPE_1'] == 'Psychic') | (my_data['TYPE_2'] == 'Psychic')]
Electric=my_data[(my_data['TYPE_1'] == 'Electric') | (my_data['TYPE_2'] == 'Electric')]
plt.scatter(Psychic.ATTACK.head(50),Psychic.DEFENSE.head(50),color='Red',label='Psychic',marker='*',s=75)
plt.scatter(Electric.ATTACK.head(50),Electric.DEFENSE.head(50),color='Blue',label='Electric',marker='o')
plt.legend()
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.plot()
plt.show()
```

```python
pie_data=pd.DataFrame(my_data.groupby(['TYPE_1'],as_index=False).count().sort_values(by='ATTACK',ascending=False).head(10))
pie_data
```

**Top 10 types of pokemons by population---Pieplot**

```python
plt.subplots(figsize=(10,10))
plt.pie(pie_data.ATTACK,labels=pie_data.TYPE_1,autopct='%1.1f%%')
plt.title('Top 10 pokemon in population')
plt.plot()
plt.show()
```
